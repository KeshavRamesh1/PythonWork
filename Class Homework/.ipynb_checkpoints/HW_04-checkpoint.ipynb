## Q4 
### 8 pts.

a. Using the stats models library, fit a simple regression of PRSM v. FICO and print its summary output.

b. Interpret the R-squared value.

c. For two loans whose FICO score differs by 100, what does the model predict their *difference* in PRSM score will be? You can do the calculation "by hand" or have Python do it. You can extract the  model coefficients from the fitted regression object using the ".params" attribute. 

d. Based on the p-value of the FICO score, does it appear that this variable is an important predictor of PRSM? Explain your reasoning.

e. Create a data frame for prediction (call it Xnew) that has a single column named "FICO" and two values, 645 and 750. Print the data frame.

f. Using the .predict method, predict and print the PRSM score for these two loans.

g. There is an attribute in the regression output called "mse_resid", the mean squared error. If you called your fitted model olsres1, then you get the MSE with the command 'olsres1.mse_resid'. Use Python to calculate the ***ROOT*** MSE.

h. Approximately how wide (the distance from the center to the edge) would a 95% prediction interval be for the predictions you made in part 4f? This is called the "margin of error" for the predictions.



```python
# Q4 code goes here
import statsmodels.formula.api as smf
import math

import statsmodels as sm
 
#Q4a.
olsmod1 = smf.ols(formula = "PRSM ~ FICO", data = prsm_data)
olsres1 = olsmod1.fit()
print(olsres1.summary())
 
#Q4b. The R-squared value, 0.252, means that approximately 25.2% of the data fits the regression model.
 
#Q4c. The model predicts that their difference in PRSM score will be approximately 0.1696
print(100 * olsres1.params[1])

#Q4d. It appears that this variable is an important predictor of PRSM because p-value = 0.000 is less than 0.05 and is therefore statistically significant.

#Q4e.
Xnew = {"FICO": [645, 750]}
print(Xnew)
 
#Q4f.
print(olsres1.predict(Xnew))
 
#Q4g.
math.sqrt(olsres1.mse_resid)
 
#Q4h. 95% confidence interval for predictions made in part 4f (or margin of error) would be approximately +/-0.41656.
print("+/-" + str(2 * (math.sqrt(olsres1.mse_resid))))
```

## Q5 Fit the interaction model to the data.

### 4 pts.

a. Fit the model that includes the variables FICO, Loan.Type and the ***interaction*** between FICO and Loan.Type. Print the model summary. There is a nuance here, in that because the variable 'Loan.Type' name contains a period, Python thinks that it should look for variable named "Loan" that has an attribute called "Type". This will cause an error! The way to fix this problem is to use a special notation in the formula where you write Q("Loan.Type") instead of just Loan.Type.

b. Based on your fitted model, which Loan.Type, Original or Repeat, is the most sensitive to FICO score, in terms of performance? That is, which of the loan types has the largest slope?

c. Is the difference in slopes statistically significant? Justify your answer.


```python
# Q5 code goes here.

#Q5a.
olsmod2 = smf.ols(formula = "FICO ~ Q('Loan.Type') + FICO:Q('Loan.Type')", data = prsm_data)
olsres2 = olsmod2.fit()
print(olsres2.summary())

#Q5b. The Loan.Type Original is most sensitive to FICO score, in terms of performance (it has the largest slope).

#Q5c. The difference in slopes is statistically significant because the p-value is less than 0.05
```

a. Create a new version of the data frame that just has the continuous variables in it. Call it 'Xcts'. 

b. Create a new version of the data frame that just has the categorical predictor variables in it. Call it "Xcat".

c. Binarize the categorical variables using the pandas get_dummies function and its 'drop_first=True' argument.

```python
# Q6 code goes here

#6a.
Xcts = prsm_data[["Commission.Upfront", "Validated.Monthly.Batch", "FICO", "Years.In.Business", "Satisfied.Accounts", "Previous.Delinquent.Credit.Lines", "Num.of.Legal.Items", "Population.in.Zip.Code", "Average.House.Value.in.Zip.Code", "Income.Per.Household.in.Zip....", "Median.Age.in.Zip.Code", "Employment.in.Zip.Code"]]
 
#6b.
Xcat = prsm_data[["Loan.Type", "Business.Entity.Type", "ISO.Name"]]
 
#6c.
Xcat = pd.get_dummies(Xcat, drop_first = True)
 
#6d.
X = pd.merge(Xcts, Xcat, left_index = True, right_index = True)
print(X.shape)
 
#6e.
print(X.head(5))
 
#6f
y = prsm_data["PRSM"]
print(y.head(5))
```
